## 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
employee_count = df["Team"].value_counts()

print(f"The number of employee in each team : \n {employee_count}")
```

### Percentage of employees with respect to the total employees


```python
employee_percentage = (employee_count / len(df)) * 100

print("The percentage of splitting : \n", employee_percentage)
```

# 2.Segregate the employees w.r.t different positions.

```python
df["Position"].value_counts()
```

```python
positions = df["Position"].unique()
for position in positions:
    print(f" Employees of {position} position : ")
    print(df.loc[df['Position'] == position, 'Name' ],"\n")
    print()
```

# 3.Find from which age group most of the employees belong to.

```python
age = df["Age"].value_counts()
print(age)
```

```python
df['Age Group'] = pd.cut(df['Age'], bins = [15, 20, 25, 30, 35, 40])
age_count = df['Age Group'].value_counts()
print(age_count)
```

```python
print(f"Most of the employees belong to the age group of : ", age_count.idxmax())
```

# 4.Find out under which team and position, spending in terms of salary is high.

```python
salary_high = df[["Team", "Position", "Salary"]].sort_values(by = "Salary", ascending = False)
salary_high 
```

```python
df.loc[df['Salary'] == df['Salary'].max(), ['Team', 'Position', 'Salary']]
```

# 5.Find if there is any correlation between age and salary , represent it visually.

```python
correlation=df['Age'].corr(df['Salary'])
print("Correlation =",correlation)
```

```python
correlation_matrix = df[['Age', 'Salary']].corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df['Age'],df['Salary'], color="blue")
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend
plt.show()
```
